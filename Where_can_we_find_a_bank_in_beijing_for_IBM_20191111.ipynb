## Custom function

```python
#Make a web link.
def call_api_a(url):
    i=0
    while True:
        i=i+1
        #print(url)
        try:
            results = requests.get(url).json()
            if results['status']=="1":
                break
            else:
                print(url,"调用接口失败..",results['info'])            
        except:
            #print(url,"调用接口失败")
            #time.sleep(1)
            print(url,"调用接口失败..",results['info'])
            time.sleep(1)
            if i==3:
                results=None
                break
            
    return results


# Get the number of results
def get_venues_num(url):   
    
    results=call_api_a(url)
    venues_num=results['count']
    
    return int(venues_num)  


# Query target information
def get_pois(url_template,key,keywords,types,city,page=1):
    
    url =url_template.format(key,keywords,types,city,page)
    
    #先计算结果总条数
    venues_num = get_venues_num(url)
    
    results = pd.DataFrame()
    if venues_num>0:
        #用总条数计算结果页数
        page_num_all=venues_num//25 + 1
        print("{} 区域内,{}-{}个数：{},总页数{}".format(city,keywords,types,venues_num,page_num_all))


        
        for num in range(page_num_all):
            #print("完成第{}页".format(num))        
            result = call_api_a(url)
            #time.sleep(1)
            venues=result['pois']
            nearby_venues = json_normalize(venues) # flatten JSON
            #nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
            results=pd.concat([results,nearby_venues],ignore_index=True)
            #print(results.shape)
        
    return results
```

```python
# Query around information
def get_around_pois(para,api_url,uid,addr):
    
    para['location']=addr
        
    para_str = urllib.parse.urlencode(para)
    url = api_url + para_str
    
    #先计算结果总条数
    venues_num = get_venues_num(url)
    
    results = pd.DataFrame()
    if venues_num>0:
        #用总条数计算结果页数
        page_num_all=venues_num//25 + 1
        print("{}-{}，米范围内,商铺个数：{},总页数{}".format(uid,para['radius'],venues_num,page_num_all))


        
        for num in range(page_num_all):
            #print("完成第{}页".format(num))        
            result = call_api_a(url)
            #time.sleep(1)
            venues=result['pois']
            nearby_venues = json_normalize(venues) # flatten JSON
            #nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
            results=pd.concat([results,nearby_venues],ignore_index=True)
            results['bank_uid']=uid
            #print(results.shape)
        
    return results



# 目标场所指定关键词 半径搜索
def get_all_around(df,para,api_url,keys_list):
    
    results = pd.DataFrame()
    for index,row in df.iterrows():
        #random.shuffle(keys_list)
        
        #para['key']= random.choice(keys_list)
        print(para['key'])
        #print(index,row['name'],row['uid'])

        #参数赋值        
        addr=row['location']
        uid=row['id']
        radius=para['radius']

        #获取结果
        result = get_around_pois(para,api_url,uid,addr)
              
        #result['目标_机构名称']=row['name']
        result['目标_uid']=uid
        #合并结果
        results=pd.concat([results,result],ignore_index=True,sort=False)   
    
        print("完成{}-{} {}范围 ,维度：{}".format(index,row['name'],radius,results.shape))
    #处理列名
    results.columns = [col.split(".")[-1] for col in results.columns] 
        
    return results
```

## Store industry category processing

First, the industry categories of the shops are split, and then the categories are merged into two categories, life and business.

```python
#拆分行业类别
def split_1(tag):
    try:
        type_str = tag.split(";")[0]
    except:
        type_str=tag
    return type_str

def split_2(tag):
    try:
        type_str = tag.split(";")[1]
    except:
        type_str=tag
    return type_str

def split_3(tag):
    try:
        type_str = tag.split(";")[2]
    except:
        type_str=tag
    return type_str
```

```python
df_banks_bj_around_pois_s['type_1']=df_banks_bj_around_pois_s['type'].apply(split_1)
df_banks_bj_around_pois_s['type_2']=df_banks_bj_around_pois_s['type'].apply(split_2)
df_banks_bj_around_pois_s['type_3']=df_banks_bj_around_pois_s['type'].apply(split_3)
```

## Artificial bank and unmanned bank

```python
# import k-means from clustering stage
from sklearn.cluster import KMeans
```

```python
bank_all_grouped_clustering = df_features.drop(columns=['bank_uid'])
```

### Determine the optimal number of classifications

```python
SSE = []  # 存放每次结果的误差平方和
for k in range(1,15):
    estimator = KMeans(n_clusters=k)  # 构造聚类器
    estimator.fit(bank_all_grouped_clustering)
    SSE.append(estimator.inertia_)
X = range(1,15)
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(X,SSE,'o-')
plt.show()
```

```python
# set number of clusters
kclusters = 4

#bank_all_grouped_clustering = df_features.drop(columns=['bank_uid'])

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=2509).fit(bank_all_grouped_clustering)
# check cluster labels generated for each row in the dataframe
#kmeans.labels_[0:10] 
all_cluster=df_features[['bank_uid']]
all_cluster.columns=['id']
all_cluster['label']=kmeans.labels_

all_cluster_3 = all_cluster
all_cluster.head()
```
