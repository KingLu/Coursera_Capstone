# P1

 scrape the following Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, 

```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
#print(soup.prettify())
```

```python
my_table = soup.find('table',{'class':'wikitable sortable'})
#my_table
```

## get Postcode

```python
Postcode = []
for row in my_table.find_all('tr')[1:]:
    state_cell = row.find_all('td')[0]  
    Postcode.append(state_cell.text)
#print(Postcode)
```

```python
## get Borough
```

```python
Borough = []
for row in my_table.find_all('tr')[1:]:
    title_cell = row.find_all('td')[1]
    #print(title_cell)
    Borough.append(title_cell.text)
#print(Borough)
```

```python
## get Neighbourhood
```

```python
Neighbourhood = []
for row in my_table.find_all('tr')[1:]:
    name_cell = row.find_all('td')[2]
    Neighbourhood.append(name_cell.text[:-1])
#print(Neighbourhood)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
df['Postcode'] = Postcode
df['Borough'] = Borough
df['Neighbourhood'] = Neighbourhood
df
```

Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
print(df.shape)
df_tor = df[df['Borough']!='Not assigned']
#df_tor
```

If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df_tor.loc[df_tor[df_tor['Neighbourhood']=='Not assigned'].index,['Neighbourhood']]=df_tor.loc[df_tor[df_tor['Neighbourhood']=='Not assigned'].index]['Borough']
```

```python
#df_tor
```

More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11 in the above table.

```python
df_tor_c =df_tor.groupby(['Postcode','Borough'])['Neighbourhood'].apply(",".join).reset_index()
df_tor_c
```

```python
print("The shape is:",df_tor_c.shape)
```

```python
import folium # map rendering library
```

```python
neighborhoods = df_tor_s
```

# create map of Toronto using latitude and longitude values

```python
latitude = 43.653908
longitude = -79.384293
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto

```
